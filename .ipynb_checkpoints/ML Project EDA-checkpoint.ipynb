```python
#Remove the variables that have more than 10% of the observations as null.
NullThreshold = 0.1
df = df.loc[:,(df.isnull().sum(axis = 0) / df.shape[0]) < NullThreshold]
```

```python
df.shape
```

```python
objcol2 = df.columns[df.dtypes == "object"]
```

```python
#Remove the variables that have more than 90% of the observations in one value.
MajorityThreshold = 0.9
for obj in objcol2:
    if (df[obj].value_counts().max() / df[obj].value_counts().sum()) > MajorityThreshold:
        df.drop(obj, axis = 1, inplace = True)
        print("Drop")
```

```python
Ordinal = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual']
NonOrdinal = ['MSZoning', 'LotShape', 'LandContour', 'LotConfig', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle',
             'RoofStyle', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation', 'BsmtExposure', 'GarageType', 'GarageFinish',
             'SaleType', 'SaleCondition']
```

```python
# Encode ordinal features
df['ExterQual'] = df['ExterQual'].replace({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex': 5})
df['ExterCond'] = df['ExterCond'].replace({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex': 5})
df['BsmtQual'] = df['BsmtQual'].replace({'NA':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex': 5})
df['BsmtFinType1'] = df['BsmtFinType1'].replace({'NA':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ': 6})
df['BsmtFinType2'] = df['BsmtFinType2'].replace({'NA':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ': 6})
df['HeatingQC'] = df['HeatingQC'].replace({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex': 5})
df['KitchenQual'] = df['KitchenQual'].replace({'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex': 5})
```

```python
# Encode non-ordinal features
labelencoder = LabelEncoder()
for obj in NonOrdinal:
    df[obj] = labelencoder.fit_transform(df[obj])
```
